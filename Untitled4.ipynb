```python
import random

#글래디에이터 정의
class Gladiator:
    def __init__(self, name, age=26):
        self.name = name            # 검투사 이름
        self.age = age          # 검투사 나이
        self.hp = age * 15        # 체력은 나이에 따라 결정 (예: 나이 20이면 체력 300)

    def attack(self, opponent):
        # 간단한 공격: 데미지는 나이와 2부터 3 사이의 랜덤 숫자를 곱해 결정합니다.
        damage = self.age * random.randint(2, 3)
        print(f"{self.name}가 {opponent.name}에게 {damage}의 데미지를 가했습니다!")
        opponent.hp -= damage

    def shield(self, opponent):
        # 방어: 데미지는 나이와 1부터 2사이의 랜덤 숫자를 곱해 결정합니다.
        depend = self.age * random.randint(2, 2)
        print(f"{opponent.name}가 {self.name}의 데미지 중 {depend}만큼 상쇄했습니다!")  
        opponent.hp += depend

    def is_knocked_out(self):
        # 체력이 0 이하이면 쓰러진 것으로 간주
        return self.hp <= 0

    def status(self):
        # 현재 상태(이름과 체력)를 출력하는 메서드
        print(f"{self.name} - 체력: {self.hp}")

# 검투사 객체 생성 (타입 정보를 제거)
최시언검투사 = Gladiator("최시언검투사", age=26)
이경훈검투사 = Gladiator("이경훈검투사", age=26)

# 배틀 시뮬레이션 시작
print("배틀 시작!")
while True:
    # 최시언이 이경훈을 공격, 최
    최시언검투사.attack(이경훈검투사)
    최시언검투사.shield(이경훈검투사) 
    if 이경훈검투사.is_knocked_out():
        print(f"{이경훈검투사.name}가 쓰러졌습니다! {최시언검투사.name} 승리!")
        break


    # 이경훈이 최시언을 공격
    이경훈검투사.attack(최시언검투사)
    이경훈검투사.shield(최시언검투사)
    if 최시언검투사.is_knocked_out():
        print(f"{최시언검투사.name}가 쓰러졌습니다! {이경훈검투사.name} 승리!")
        break

    # 각 검투사의 현재 체력을 출력
    최시언검투사.status()
    이경훈검투사.status()
    print("----------")
```
